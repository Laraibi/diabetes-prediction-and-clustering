```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
DATA_PATH = Path("../data/diabetes.csv")

df = pd.read_csv(DATA_PATH)

df.head()
```

```python
df = df.drop(columns=["Unnamed: 0"])

df.head()
```

```python
print("Dataset shape:")
print(df.shape)

print("\nColumns:")
print(df.columns.tolist())

print("\nData types:")
print(df.dtypes)
```

```python
df.isnull().sum()
```

```python
columns_with_possible_missing_values = [
    "Glucose",
    "BloodPressure",
    "SkinThickness",
    "Insulin",
    "BMI"
]

for column in columns_with_possible_missing_values:
    zero_count = (df[column] == 0).sum()

    print(f"{column}: {zero_count} zero values")
```

```python
df.duplicated().sum()
```

```python
df.describe()
```

```python
df.hist(figsize=(15, 10))

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(10, 8))

sns.heatmap(
    df.corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")

plt.show()
```
